# 08. BigQuery SQL 분석

## 목표
- BigQuery에 적재된 원본 데이터(transactions, cards, users)를 대상으로 SQL 분석
- SQLite에서 수행한 분석을 BigQuery 문법으로 재현
- 분석 결과를 View로 저장 → Looker Studio 연동 기반 마련

## SQLite vs BigQuery 주요 문법 차이
| 항목 | SQLite | BigQuery |
|---|---|---|
| 문자열 → 숫자 변환 | `CAST(REPLACE(Amount, '$', '') AS FLOAT)` | `SAFE_CAST(REPLACE(Amount, '$', '') AS FLOAT64)` |
| 뷰 생성 | `CREATE VIEW IF NOT EXISTS` | `CREATE OR REPLACE VIEW` |
| 컬럼명 특수문자 | `"Errors?"` | 백틱으로 `` `Errors_` `` (V2 자동 변환됨) |
| 서브쿼리 별칭 | 별칭 없어도 됨 | 반드시 별칭 필요 |

## 진행 방식

Python 노트북에서 BigQuery API를 직접 호출하는 대신,
**BigQuery 콘솔에서 SQL을 직접 실행**하고 쿼리는 `sql/` 폴더에 파일로 저장하는 방식으로 진행한다.

### 이유
- 패키지 의존성 문제 없이 BigQuery 본연의 실행 환경 활용
- SQL 파일을 깃허브에 올려 코드 기록 유지
- View를 생성해두면 Looker Studio에서 바로 연동 가능

### 파일 구성
| 파일 | 내용 |
|---|---|
| `sql/bigquery_views.sql` | 분석용 View 생성 쿼리 (clean_transactions, ltv_transactions) |
| `sql/bigquery_analysis.sql` | EDA · AARRR · Cohort · LTV · RFM 분석 쿼리 |

### 실행 순서
1. `sql/bigquery_views.sql` → BigQuery 콘솔에서 실행하여 View 생성
2. `sql/bigquery_analysis.sql` → BigQuery 콘솔에서 분석 쿼리 실행
3. View 기반으로 Looker Studio 연동 (`09_looker_studio.ipynb` 에서 연동 확인)
